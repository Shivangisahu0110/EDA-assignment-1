Q1. What are the key features of the wine quality data set? Discuss the importance of each feature in
predicting the quality of wine.

```python
import pandas as pd 
df = pd.read_csv('winequality-red.csv',sep = ';')
df.info()
```

Here are the key features of the Wine Quality dataset:

Fixed Acidity: This feature represents the concentration of non-volatile acids (like tartaric acid) in the wine. Higher fixed acidity contributes to the tartness and sharpness of the wine. It is important in maintaining a balanced pH in wine.

Volatile Acidity: This refers to the amount of acetic acid in wine, too much of which can lead to an unpleasant vinegar taste. Volatile acidity is a crucial factor in the quality of wine, as high levels can lead to a wine being described as having 'vinegar' or 'nail polish' defects.

Citric Acid: Found naturally in small quantities in grapes, citric acid is sometimes added to wines to increase acidity, enhance flavor, and act as a preservative. It influences the freshness of the wine.

Residual Sugar: This measures the amount of sugar remaining after fermentation stops. It affects the sweetness of the wine. Wines can range from dry to sweet, depending on their residual sugar levels.

Chlorides: The amount of salts (primarily sodium chloride) in the wine. High levels of chlorides can lead to a salty taste, which is generally not desirable in wines.

Free Sulfur Dioxide (SO2): This is the portion of sulfur dioxide that is not bound to other molecules in wine and is effective in preventing microbial growth and the oxidation of wine. It acts as a preservative.

Total Sulfur Dioxide: This represents the total amount of bound and free forms of S02. In high concentrations, sulfur dioxide can give off an unwanted burnt match smell and can affect the respiratory system of sensitive tasters.

Density: The density of wine is close to that of water depending on the alcohol and sugar content. It provides insights into the alcohol and sugar content of the wine.

pH: A measure of the acidity or basicity of the wine on a logarithmic scale. Most wines are slightly acidic and pH can influence color, stability, and taste.

Sulphates: These are compounds that contribute to the levels of sulfur dioxide gas in the wine, acting as antimicrobials and antioxidants. It is crucial in preventing oxidation and maintaining freshness.

Alcohol: The percentage of alcohol present in the wine. Generally, higher alcohol content can lead to a perception of richness and body in wine.

Quality (Target Variable): This is typically scored between 0 (very bad) and 10 (very excellent). The quality score is the target variable that models attempt to predict based on the chemical properties listed above.

## insight = there is no missing values in dataset

Some Imputation Techniques

Deleting Rows

Advantage: Keeps the dataset pure from any artificial data which could potentially bias the model.

Disadvantage: Loss of valuable data, especially if missingness is extensive or not random. This could reduce the statistical power of the model.

Mean/Median/Mode Imputation

Advantage: Simple and fast to implement; works well with numerical data when the data is normally distributed (mean) or in the presence of outliers (median).

Disadvantage: Can reduce the variance of the dataset; assumes that the missing values are randomly distributed which might not always be the case.

K-Nearest Neighbors (KNN)

Advantage: Takes into account the similarity between instances, potentially leading to more realistic imputations.

Disadvantage: Computationally expensive; performance depends on the choice of 'k' and the distance metric; not ideal for high-dimensional data due to the "curse of dimensionality".

Regression Imputation

Advantage: Utilizes the correlations between features for predicting the missing values. Can be more accurate than mean/median/mode if relationships in the data are strong.

Disadvantage: Risk of creating artificial data that can lead to model overfitting; assumes linear relationships between features.

Multiple Imputation

Advantage: Accounts for the uncertainty in the imputations by creating multiple complete datasets and averaging results; suitable for handling missing data not at random (MNAR).

Disadvantage: More complex and time-consuming; requires sophisticated statistical software and expertise.

Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of
features. What is the minimum number of principal components required to explain 90% of the variance in
the data?

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Step 1: Load the data
df = pd.read_csv('winequality-red.csv',sep = ';')

# Step 2: Preprocess the data
features = df.drop('quality', axis=1)

# Standardize the features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Step 3: Apply PCA
pca = PCA()
components = pca.fit_transform(features_scaled)

# Step 4: Determine the number of components
explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)
num_components = np.where(cumulative_variance >= 0.90)[0][0] + 1  # Indexing starts at 0

# Plotting the explained variance
plt.figure(figsize=(10, 5))
plt.plot(cumulative_variance, marker='o')
plt.title('Cumulative Explained Variance by PCA Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.axhline(y=0.90, color='r', linestyle='--')
plt.axvline(x=num_components-1, color='r', linestyle='--')
plt.grid(True)
plt.show()

print(f"Minimum number of principal components required to explain 90% of the variance: {num_components}")
```
